```python
# 1. –í—ã–±–µ—Ä–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –ø–æ –Ω–æ–º–µ—Ä—É –≤–∞—Ä–∏–∞–Ω—Ç–∞.
# 2. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏. –ì—Ä–∞—Ñ–∏–∫ –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å –∑–∞–≥–æ–ª–æ–≤–æ–∫ –∏ –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π.
# 3. –°–∫–æ–ø–∏—Ä—É–π—Ç–µ –∫–æ–¥ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –∏–∑–º–µ–Ω–∏—Ç–µ –≤–Ω–µ—à–Ω–∏–π –≤–∏–¥ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–µ–π, –∏—Å–ø–æ–ª—å–∑—É—è –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã.
# 4. –ò—Å–ø–æ–ª—å–∑—É—è –∞–Ω–Ω–æ—Ç–∞—Ü–∏—é ‚Äì –≤—ã–¥–µ–ª–∏—Ç–µ: –ª–æ–∫–∞–ª—å–Ω—ã–π –º–∏–Ω–∏–º—É–º (–∏–ª–∏ –º–∞–∫—Å–∏–º—É–º) —Ñ—É–Ω–∫—Ü–∏–∏, —Ç–æ—á–∫—É —Ä–∞–∑—Ä—ã–≤–∞, —Ç–æ—á–∫—É –ø–µ—Ä–µ–≥–∏–±–∞. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ä–∞–∑–Ω—ã–µ —Ç–∏–ø—ã —Å—Ç—Ä–µ–ª–æ–∫.
# 5. –î–æ–±–∞–≤—å—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –Ω–∞–¥–ø–∏—Å—å (—Ç–µ–∫—Å—Ç –ø—Ä–∏–¥—É–º–∞–π—Ç–µ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ). 
# 6. –î–æ–±–∞–≤—å—Ç–µ –µ—â–µ –æ–¥–∏–Ω –≥—Ä–∞—Ñ–∏–∫ (–ª—é–±–æ–π), –æ—Ñ–æ—Ä–º–∏—Ç–µ –ª–µ–≥–µ–Ω–¥—É.
```

```python
import numpy as np
import matplotlib.pyplot as plt
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
def fun(x):
    return np.sqrt(3)*x-2*np.sin(x)

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–π x –æ—Ç 0 –¥–æ œÄ —Å —à–∞–≥–æ–º 0.01
x_values = np.arange(-2*np.pi, 2*np.pi, 0.1)
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π y
y=fun(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8,6))
plt.plot(x_values, y, label='$y(x) = \sqrt{3}x - 2\sin(x)$')
plt.title("–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y(x) = \sqrt{3}x - 2\sin(x)$")
plt.xlabel('x')
plt.ylabel('y')
```

#### –°–∫–æ–ø–∏—Ä—É–π—Ç–µ –∫–æ–¥ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –∏–∑–º–µ–Ω–∏—Ç–µ –≤–Ω–µ—à–Ω–∏–π –≤–∏–¥ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–µ–π, –∏—Å–ø–æ–ª—å–∑—É—è –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã.

```python
import numpy as np
import matplotlib.pyplot as plt
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
def fun(x):
    return np.sqrt(3)*x-2*np.sin(x)

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–π x –æ—Ç 0 –¥–æ œÄ —Å —à–∞–≥–æ–º 0.01
x_values = np.arange(-2*np.pi, 2*np.pi, 0.1)
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π y
y=fun(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8,6))
plt.plot(x_values, y, label='$y(x) = \sqrt{3}x - 2\sin(x)$')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y(x) = \sqrt{3}x - 2\sin(x)$', fontsize=16, fontweight='bold', color='blue', loc='left', pad=20)
plt.xlabel('X\n–ù–µ–∑–∞–≤–∏—Å–∏–º–∞—è –≤–µ–ª–∏—á–∏–Ω–∞', fontsize=14, fontweight='bold', color='green', labelpad=10)
plt.ylabel('Y\n–ó–∞–≤–∏—Å–∏–º–∞—è –≤–µ–ª–∏—á–∏–Ω–∞', fontsize=14, fontweight='bold', color='red', labelpad=10)

# –ü–æ–∏—Å–∫ –ª–æ–∫–∞–ª—å–Ω—ã—Ö —ç–∫—Å—Ç—Ä–µ–º—É–º–æ–≤
local_max_index = np.argmax(y)
local_min_index = np.argmin(y)
plt.annotate('–õ–æ–∫–∞–ª—å–Ω—ã–π –º–∞–∫—Å–∏–º—É–º', xy=(x_values[local_max_index], y[local_max_index]), xytext=(-3, 2),
             arrowprops=dict(facecolor='pink',arrowstyle='simple'))
plt.annotate('–õ–æ–∫–∞–ª—å–Ω—ã–π –º–∏–Ω–∏–º—É–º', xy=(x_values[local_min_index], y[local_min_index]), xytext=(1, -2),
             arrowprops=dict(arrowstyle='->'))

# –ü–æ–∏—Å–∫ —Ç–æ—á–µ–∫ —Ä–∞–∑—Ä—ã–≤–∞ –∏ –ø–µ—Ä–µ–≥–∏–±–∞
discontinuity_points = [-np.pi, 0, np.pi]
inflection_points = [0]
plt.axvline(x=0, color='gray', linestyle='--')
plt.annotate('–¢–æ—á–∫–∞ —Ä–∞–∑—Ä—ã–≤–∞', xy=(0, fun(0)), xytext=(-6, -1),
             arrowprops=dict(facecolor='green', arrowstyle='fancy'))
plt.annotate('–¢–æ—á–∫–∞ –ø–µ—Ä–µ–≥–∏–±–∞', xy=(0, fun(0)), xytext=(4, 2),
             arrowprops=dict(facecolor='purple', arrowstyle='-|>'))

 # –î–æ–±–∞–≤—å—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –Ω–∞–¥–ø–∏—Å—å (—Ç–µ–∫—Å—Ç –ø—Ä–∏–¥—É–º–∞–π—Ç–µ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ). 
plt.text(-6,6,'nihao',fontsize=12, bbox=dict(facecolor='lightgray', alpha=0.5))

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤—Ç–æ—Ä–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –ª–µ–≥–µ–Ω–¥—ã
plt.plot(x_values, np.cos(x_values), label='$y = \cos(x)$', linestyle='--')
plt.legend()
```
