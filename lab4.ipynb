```python
# 1. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ stairs-–≥—Ä–∞—Ñ–∏–∫ (—Å—Ç—É–ø–µ–Ω—á–∞—Ç—ã–π) –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ y(x). –ü—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–∏–¥
# —Ñ—É–Ω–∫—Ü–∏–π –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤—ã–±—Ä–∞—Ç—å —Å–æ–≥–ª–∞—Å–Ω–æ –≤–∞—Ä–∏–∞–Ω—Ç—É. –¢–æ–ª—â–∏–Ω—É –ª–∏–Ω–∏–∏, —Ä–∞–∑–º–µ—Ä –∏ —Ü–≤–µ—Ç
# –º–∞—Ä–∫–µ—Ä–∞, –∞ —Ç–∞–∫–∂–µ –∫–æ–Ω—Ç—É—Ä–∞ –º–∞—Ä–∫–µ—Ä–∞ –ø–æ–¥–æ–±—Ä–∞—Ç—å —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ.
# ùë¶(ùë•) = 1 ‚àí \sqrt[3]{x^2}   x‚àà[0,10], ‚àÜùë• = 0,6
```

```python
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ y(x)
def y(x):
    return 1 - np.cbrt(x ** 2)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –∑–Ω–∞—á–µ–Ω–∏–π x
x_values = np.arange(0, 10, 0.6)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π y –¥–ª—è –∫–∞–∂–¥–æ–≥–æ x
y_values = y(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Å—Ç—É–ø–µ–Ω—á–∞—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8, 6))
plt.step(x_values, y_values, where='mid', linewidth=2, color='red', linestyle='-.', marker='s', markersize=8, markerfacecolor='white', markeredgewidth=1, markeredgecolor='black', label='$y(x) = 1 - \\sqrt[3]{x^2}$')

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π –∫ –æ—Å—è–º
plt.title('–°—Ç—É–ø–µ–Ω—á–∞—Ç—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y(x) = 1 - \\sqrt[3]{x^2}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å–µ—Ç–∫–∏
plt.grid(True)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ª–µ–≥–µ–Ω–¥—ã
plt.legend()

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.show()
```

```python
# 2. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ stem-–≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ y(x), ùë• ‚àà [0,2ùúã], ‚àÜùë• = 0,2. –ü—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏
# –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–∏–¥ —Ñ—É–Ω–∫—Ü–∏–π –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤—ã–±—Ä–∞—Ç—å —Å–æ–≥–ª–∞—Å–Ω–æ –≤–∞—Ä–∏–∞–Ω—Ç—É. –¢–æ–ª—â–∏–Ω—É –ª–∏–Ω–∏–∏,
# —Ä–∞–∑–º–µ—Ä –∏ —Ü–≤–µ—Ç –º–∞—Ä–∫–µ—Ä–∞, –∞ —Ç–∞–∫–∂–µ –∫–æ–Ω—Ç—É—Ä–∞ –º–∞—Ä–∫–µ—Ä–∞ –ø–æ–¥–æ–±—Ä–∞—Ç—å —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ.

#y(x)=6sin(e^x)   –ë–∏—Ä—é–∑–æ–≤—ã–π, –ø—è—Ç–∏–∫–æ–Ω–µ—á–Ω–∞—è –∑–≤–µ–∑–¥–∞, —à—Ç—Ä–∏—Ö-–ø—É–Ω–∫—Ç–∏—Ä–Ω–∞—è
```

```python
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ y(x)
def y_function(x):
    return 6 * np.sin(np.exp(x))

# –ó–∞–¥–∞–Ω–∏–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ x
x_values = np.arange(0, 2*np.pi + 0.2, 0.2)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π y –¥–ª—è –∫–∞–∂–¥–æ–≥–æ x
y_values = y_function(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ stem-–≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8, 6))  # –†–∞–∑–º–µ—Ä –≥—Ä–∞—Ñ–∏–∫–∞

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –º–∞—Ä–∫–µ—Ä–æ–≤
markerline, stemlines, baseline = plt.stem(x_values, y_values, linefmt='C0-.', markerfmt='p', basefmt=' ')

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –º–∞—Ä–∫–µ—Ä–æ–≤
plt.setp(stemlines, linewidth=2, color='Cyan')  # –¢–æ–ª—â–∏–Ω–∞ –ª–∏–Ω–∏–π —Å—Ç–µ–±–ª–µ–π –∏ –∏—Ö —Ü–≤–µ—Ç
plt.setp(markerline, markersize=10)  # –†–∞–∑–º–µ—Ä –º–∞—Ä–∫–µ—Ä–æ–≤

# –°–∫—Ä—ã—Ç—å –±–∞–∑–æ–≤—É—é –ª–∏–Ω–∏—é
plt.setp(baseline, visible=False)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–µ–π
plt.title(r'Stem-–≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y(x) = 6\sin(e^x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# –î–∞–Ω–Ω—ã–µ Y (—É—Ä–æ–∂–∞–π–Ω–æ—Å—Ç—å) –∏ X (—ç–Ω–µ—Ä–≥–æ–æ–±–µ—Å–ø–µ—á–µ–Ω–Ω–æ—Å—Ç—å)
Y = [12.2, 8.1, 6.4, 11.3, 5.7, 12.1, 13.9, 18.5, 19.3, 13.1, 33, 9.6, 10.7, 9.9, 11.6, 12.0, 8.9, 15.5, 22.4, 16.6]
X = [10186, 7020, 11807, 11869, 3494, 2430, 27080, 5574, 1181, 8085, 4829, 3822, 9953, 4702, 12777, 12531, 4829, 8144, 6000, 2640]

# –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º (–ø–æ 5 –≥–æ–¥–∞ –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ)
num_periods = len(Y) // 5
Y_groups = [Y[i*5:(i+1)*5] for i in range(num_periods)]
X_groups = [X[i*5:(i+1)*5] for i in range(num_periods)]

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã
Y_means = [np.mean(group) for group in Y_groups]
X_means = [np.mean(group) for group in X_groups]

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Å—Ç–æ–ª–±—á–∞—Ç—ã—Ö –¥–∏–∞–≥—Ä–∞–º–º
plt.figure(figsize=(12, 6))

# –î–∏–∞–≥—Ä–∞–º–º–∞ –¥–ª—è —Å—Ä–µ–¥–Ω–µ–≥–æ Y –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º
plt.subplot(1, 2, 1)  # –ü–µ—Ä–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ (–ª–µ–≤–∞—è —á–∞—Å—Ç—å)
plt.bar(np.arange(num_periods), Y_means, color='blue', alpha=0.7)
plt.title('–°—Ä–µ–¥–Ω—è—è —É—Ä–æ–∂–∞–π–Ω–æ—Å—Ç—å –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º')
plt.xlabel('–ü—è—Ç–∏–ª–µ—Ç–∫–∞')
plt.ylabel('–°—Ä–µ–¥–Ω—è—è —É—Ä–æ–∂–∞–π–Ω–æ—Å—Ç—å (Y)')
plt.xticks(np.arange(num_periods), ['1-—è', '2-—è', '3-—è', '4-—è'])  # –ù–∞–∑–≤–∞–Ω–∏—è –ø—è—Ç–∏–ª–µ—Ç–æ–∫

# –î–∏–∞–≥—Ä–∞–º–º–∞ –¥–ª—è —Å—Ä–µ–¥–Ω–µ–≥–æ X –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º
plt.subplot(1, 2, 2)  # –í—Ç–æ—Ä–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ (–ø—Ä–∞–≤–∞—è —á–∞—Å—Ç—å)
plt.bar(np.arange(num_periods), X_means, color='green', alpha=0.7)
plt.title('–°—Ä–µ–¥–Ω—è—è —ç–Ω–µ—Ä–≥–æ–æ–±–µ—Å–ø–µ—á–µ–Ω–Ω–æ—Å—Ç—å –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º')
plt.xlabel('–ü—è—Ç–∏–ª–µ—Ç–∫–∞')
plt.ylabel('–°—Ä–µ–¥–Ω—è—è —ç–Ω–µ—Ä–≥–æ–æ–±–µ—Å–ø–µ—á–µ–Ω–Ω–æ—Å—Ç—å (X)')
plt.xticks(np.arange(num_periods), ['1-—è', '2-—è', '3-—è', '4-—è'])  # –ù–∞–∑–≤–∞–Ω–∏—è –ø—è—Ç–∏–ª–µ—Ç–æ–∫

# –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –¥–∏–∞–≥—Ä–∞–º–º—ã
plt.tight_layout()  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–∞–∑–º–µ—â–µ–Ω–∏—è
plt.show()
```

```python
import matplotlib.pyplot as plt

# –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ Y
Y = [12.2, 8.1, 6.4, 11.3, 5.7, 12.1, 13.9, 18.5, 19.3, 13.1, 33, 9.6, 10.7, 9.9, 11.6, 12.0, 8.9, 15.5, 22.4, 16.6]

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –ø—è—Ç–∏–ª–µ—Ç–∫–∏
five_years = Y[0:5]
ten_years = Y[5:10]
fifteen_years = Y[10:15]
twenty_years = Y[15:20]

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—É–º–º –¥–ª—è –∫–∞–∂–¥–æ–π –ø—è—Ç–∏–ª–µ—Ç–∫–∏
sum_five = sum(five_years)
sum_ten = sum(ten_years)
sum_fifteen = sum(fifteen_years)
sum_twenty = sum(twenty_years)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–≥–æ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è
total_sum = sum(Y)
percentages = [sum_five/total_sum*100, sum_ten/total_sum*100, sum_fifteen/total_sum*100, sum_twenty/total_sum*100]

# –ú–µ—Ç–∫–∏ –¥–ª—è –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
labels = ['1-—è', '2-—è', '3-—è', '4-—è']

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–æ –ø—è—Ç–∏–ª–µ—Ç–∫–∞–º')
plt.axis('equal')  # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–≤–Ω—ã—Ö –º–∞—Å—à—Ç–∞–±–æ–≤ –ø–æ –æ—Å—è–º –¥–ª—è –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
plt.show()
```
