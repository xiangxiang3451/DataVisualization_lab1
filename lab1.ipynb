```python
#1. Постройте график функции y(x)
#y(x) = sin^2(3x) * sin(10x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Определение функции y(x) = sin^2(3x) * sin(10x)
def y(x):
    return (np.sin(3*x))**2 * np.sin(10*x)
# Определение диапазона значений x от 0 до π с шагом 0.01
x_values = np.arange(0, np.pi, 0.01)

# Вычисление значений y для каждого значения x
y_values = [y(x) for x in x_values]

# Построение графика
plt.figure(figsize=(10, 6))  # Создание объекта "figure" для графика размером 10x6 дюймов
plt.plot(x_values, y_values, label='$y(x) = \sin^2(3x) \cdot \sin(10x)$')  # Построение графика функции
plt.xlabel('x')  # Установка подписи оси x
plt.ylabel('y')  # Установка подписи оси y
plt.title('График функции $y(x) = \sin^2(3x) \cdot \sin(10x)$')  # Установка заголовка графика
plt.grid(True)  # Включение сетки на графике
plt.show()  # Отображение графика
```

```python
#2. Постройте в одном окне графики двух функции y1(x) и y2(x)
#𝑦1(𝑥) = 10√2𝑥 ,𝑦2(𝑥) = sin(10𝑥),𝑥 ∈ [0,4], ∆𝑥 = 0,02
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,4,0.02)

y1=10*np.sqrt(2*x)
y2=np.sin(10*x)

plt.figure(figsize=(10,6))
plt.plot(x,y1,label='$y_1(x) = 10\sqrt{2x}$', color='blue')
plt.plot(x, y2, label='$y_2(x) = \sin(10x)$', color='red')

plt.title('Графики функций $y_1(x)$ и $y_2(x)$')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
# Отображение графиков
plt.grid(True)
plt.show()
```

```python
# 3. Добавьте к указанным двум графикам третий – по своему усмотрению. И постройте эти
# три графика в разделенных окнах:
# • Используя функцию subplot();
# • Используя функцию subplots();
# • Свободная компоновка с GridSpec
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Создание массива значений x
x = np.arange(0, 4, 0.02)

# Функции y1(x),y2(x),y3(x)
y1 = 10 * np.sqrt(2 * x)
y2 = np.sin(10 * x)
y3 = np.cos(5 * x)  # Добавленная третья функция 

# Построение графиков с помощью subplot()
plt.figure(figsize=(15, 5))

# График 1
plt.subplot(1, 3, 1)
plt.plot(x, y1, label='$y_1(x) = 10\sqrt{2x}$', color='blue')
plt.title('График $y_1(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# График 2
plt.subplot(1, 3, 2)
plt.plot(x, y2, label='$y_2(x) = \sin(10x)$', color='red')
plt.title('График $y_2(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# График 3
plt.subplot(1, 3, 3)
plt.plot(x, y3, label='$y_3(x) = \cos(5x)$', color='green')
plt.title('График $y_3(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()  # Подстройка расположения графиков
plt.show()

# Построение графиков с помощью subplots()
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# График 1
axs[0].plot(x, y1, label='$y_1(x) = 10\sqrt{2x}$', color='blue')
axs[0].set_title('График $y_1(x)$')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()
axs[0].grid(True)

# График 2
axs[1].plot(x, y2, label='$y_2(x) = \sin(10x)$', color='red')
axs[1].set_title('График $y_2(x)$')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend()
axs[1].grid(True)

# График 3
axs[2].plot(x, y3, label='$y_3(x) = \cos(5x)$', color='green')
axs[2].set_title('График $y_3(x)$')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()

# Построение графиков с помощью GridSpec
fig = plt.figure(figsize=(15, 5))
gs = GridSpec(1, 3, figure=fig)

# График 1
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(x, y1, label='$y_1(x) = 10\sqrt{2x}$', color='blue')
ax1.set_title('График $y_1(x)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

# График 2
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(x, y2, label='$y_2(x) = \sin(10x)$', color='red')
ax2.set_title('График $y_2(x)$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid(True)

# График 3
ax3 = fig.add_subplot(gs[0, 2])
ax3.plot(x, y3, label='$y_3(x) = \cos(5x)$', color='green')
ax3.set_title('График $y_3(x)$')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
