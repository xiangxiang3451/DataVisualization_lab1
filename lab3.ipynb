```python
# 1.–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≤ –æ–¥–Ω–æ–º –æ–∫–Ω–µ –≥—Ä–∞—Ñ–∏–∫–∏ –¥–≤—É—Ö —Ñ—É–Ω–∫—Ü–∏–∏ y1(x) –∏ y2(x), ùë• ‚àà [0,2ùúã], ‚àÜùë• = 0,2. –ü—Ä–∏
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–∏–¥ —Ñ—É–Ω–∫—Ü–∏–π –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤—ã–±—Ä–∞—Ç—å —Å–æ–≥–ª–∞—Å–Ω–æ –≤–∞—Ä–∏–∞–Ω—Ç—É. –¢–æ–ª—â–∏–Ω—É
# –ª–∏–Ω–∏–∏, —Ä–∞–∑–º–µ—Ä –∏ —Ü–≤–µ—Ç –º–∞—Ä–∫–µ—Ä–∞, –∞ —Ç–∞–∫–∂–µ –∫–æ–Ω—Ç—É—Ä–∞ –º–∞—Ä–∫–µ—Ä–∞ –ø–æ–¥–æ–±—Ä–∞—Ç—å —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ.
# 2. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –æ–¥–∏–Ω –∏–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –≤–∞—à–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ (y1 –∏–ª–∏ y2), –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –¥–ª—è
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –Ω–µ –º–µ–Ω–µ–µ 500 —Ç–æ—á–µ–∫.
# 3. –ü–æ–¥–±–µ—Ä–∏—Ç–µ –Ω–∞–∏–ª—É—á—à–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –º–∞—Ä–∫–∏—Ä–æ–≤–∫–∏ (–Ω–∞ –æ—Å–Ω–æ–≤–µ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
# –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤).
# 4. –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Å –∑–∞–ª–∏–≤–∫–æ–π (2-3 –≤–∞—Ä–∏–∞–Ω—Ç–∞).
# 5. –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫–∏ –æ–±—Ä–µ–∑–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞ (–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –¥–ª—è –æ–±—Ä–µ–∑–∫–∏ –ø—Ä–∏–¥—É–º–∞–π—Ç–µ
# —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ).
# y1(x)= ‚àöùë•^3,  y2(x)= 2,5ùë†ùëñùëõ^2(ùë•)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# –ó–∞–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏–∏
def y1(x):
    return np.sqrt(x**3)

def y2(x):
    return  2.5*np.sin(x)**2

# –ó–∞–¥–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è x
x=np.arange(0,2*np.pi,0.2)

plt.figure(figsize=(10,6))

# –ì—Ä–∞—Ñ–∏–∫ y1
plt.plot(x, y1(x), label='$y_1(x)=\sqrt{x^3}$', color='pink', linewidth=1, linestyle='--')

# –ì—Ä–∞—Ñ–∏–∫ y2
plt.plot(x, y2(x), label='$y_2(x)=2.5sin^2(x)$', color='lime', linewidth=1, linestyle='-.')
# –î–æ–±–∞–≤–ª—è–µ–º –º–∞—Ä–∫–µ—Ä—ã
plt.scatter(x, y1(x), color='pink', s=20, marker='o', edgecolors='black')
plt.scatter(x, y2(x), color='lime', s=40, marker='x', edgecolors='black')
# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.xlabel('x')
plt.ylabel('y')
plt.title('–ì—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π $y_1(x)$ –∏ $y_2(x)$')
plt.legend()
plt.grid(True)
```

```python
# 2. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –æ–¥–∏–Ω –∏–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –≤–∞—à–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ (y1 –∏–ª–∏ y2), –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –¥–ª—è
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –Ω–µ –º–µ–Ω–µ–µ 500 —Ç–æ—á–µ–∫.
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é y1(x)
def y1(x):
    return np.sqrt(x**3)

# –°–æ–∑–¥–∞–µ–º 500 —Ç–æ—á–µ–∫ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 2*pi
x_values = np.arange(0.0, 5, 0.01)

# –í—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è y1 –¥–ª—è –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏ x
y_values = y1(x_values)

# –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$y_1(x)=\sqrt{x^3}$', color='blue')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y_1(x)=\sqrt{x^3}$ —Å 500 —Ç–æ—á–∫–∞–º–∏')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. –ü–æ–¥–±–µ—Ä–∏—Ç–µ –Ω–∞–∏–ª—É—á—à–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –º–∞—Ä–∫–∏—Ä–æ–≤–∫–∏ (–Ω–∞ –æ—Å–Ω–æ–≤–µ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
# –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤.
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª–∏–º —Ñ—É–Ω–∫—Ü–∏—é y1(x)
def y1(x):
    return np.sqrt(x**3)

# –°–æ–∑–¥–∞–¥–∏–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
x_values = np.linspace(0, 2*np.pi, 50)
y_values = y1(x_values)

# –ó–∞–¥–∞–¥–∏–º —Ä–∞–∑–º–µ—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(15, 5))

# –°—Ç–∏–ª—å 1: –∫—Ä—É–≥–ª—ã–µ –º–∞—Ä–∫–µ—Ä—ã, —Å–∏–Ω–∏–π —Ü–≤–µ—Ç
plt.subplot(1, 3, 1)
plt.plot(x_values, y_values, marker='o', color='blue', linestyle='', label='–°—Ç–∏–ª—å 1')
plt.title('–°—Ç–∏–ª—å 1')
plt.legend()

# –°—Ç–∏–ª—å 2: –∫—Ä–µ—Å—Ç–∏–∫–æ–≤—ã–µ –º–∞—Ä–∫–µ—Ä—ã, –∫—Ä–∞—Å–Ω—ã–π —Ü–≤–µ—Ç
plt.subplot(1, 3, 2)
plt.plot(x_values, y_values, marker='x', color='red', linestyle='', label='–°—Ç–∏–ª—å 2')
plt.title('–°—Ç–∏–ª—å 2')
plt.legend()

# –°—Ç–∏–ª—å 3: —Ç—Ä–µ—É–≥–æ–ª—å–Ω—ã–µ –º–∞—Ä–∫–µ—Ä—ã, –∑–µ–ª–µ–Ω—ã–π —Ü–≤–µ—Ç
plt.subplot(1, 3, 3)
plt.plot(x_values, y_values, marker='^', color='green', linestyle='', label='–°—Ç–∏–ª—å 3')
plt.title('–°—Ç–∏–ª—å 3')
plt.legend()

# –ü–æ–¥–≥–æ–Ω—è–µ–º –∫–æ–º–ø–æ–Ω–æ–≤–∫—É
plt.tight_layout()

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.show()
```

```python
# 4. –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Å –∑–∞–ª–∏–≤–∫–æ–π (2-3 –≤–∞—Ä–∏–∞–Ω—Ç–∞).
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é y1(x)
def y1(x):
    return np.sqrt(x**3)

# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ
x_values = np.linspace(0, 2*np.pi, 100)
y_values = y1(x_values)

# –°–æ–∑–¥–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(10, 6))

# –í–∞—Ä–∏–∞–Ω—Ç 1: –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –≤—Å–µ–π –æ–±–ª–∞—Å—Ç–∏ –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º –æ–¥–Ω–∏–º —Ü–≤–µ—Ç–æ–º
plt.subplot(2, 2, 1)
plt.plot(x_values, y_values, color='blue', label='–§—É–Ω–∫—Ü–∏—è y1(x)')
plt.fill_between(x_values, y_values, color='lightblue', alpha=0.3)
plt.title('–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –≤—Å–µ–π –æ–±–ª–∞—Å—Ç–∏')
plt.legend()

# –í–∞—Ä–∏–∞–Ω—Ç 2: –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –¥–≤—É—Ö —Ä–∞–∑–Ω—ã—Ö —Ü–≤–µ—Ç–æ–≤ –¥–ª—è –∑–∞–ª–∏–≤–∫–∏ –º–µ–∂–¥—É –≥—Ä–∞—Ñ–∏–∫–æ–º –∏ –æ—Å—å—é X
plt.subplot(2, 2, 2)
plt.plot(x_values, y_values, color='red', label='–§—É–Ω–∫—Ü–∏—è y1(x)')
plt.fill_between(x_values, y_values, color='pink', alpha=0.3)
plt.title('–ó–∞–ª–∏–≤–∫–∞ –º–µ–∂–¥—É –≥—Ä–∞—Ñ–∏–∫–æ–º –∏ –æ—Å—å—é X')
plt.legend()

# –í–∞—Ä–∏–∞–Ω—Ç 3: –ó–∞–ª–∏–≤–∫–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º, —á—Ç–æ–±—ã –≤—ã–¥–µ–ª–∏—Ç—å –∏–ª–∏ –ø–æ–¥—á–µ—Ä–∫–Ω—É—Ç—å –¥–∞–Ω–Ω—ã–µ
plt.subplot(2, 2, 3)
plt.plot(x_values, y_values, color='green', label='–§—É–Ω–∫—Ü–∏—è y1(x)')
plt.fill_between(x_values, y_values, where=(x_values < np.pi), color='lightgreen', alpha=0.3)
plt.title('–ó–∞–ª–∏–≤–∫–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 5. –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫–∏ –æ–±—Ä–µ–∑–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞ (–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –¥–ª—è –æ–±—Ä–µ–∑–∫–∏ –ø—Ä–∏–¥—É–º–∞–π—Ç–µ
# —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ).import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt

# –°–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–∏–º–µ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
x = np.linspace(0, 10, 100)
y = np.sin(x)

y_masked=np.ma.masked_where(y<0,y)
plt.ylim(-1,1)
plt.plot(x,y_masked,linewidth=3)
```
